```python
df_traders['Timestamp IST'] = pd.to_datetime(df_traders['Timestamp IST'], format='%d-%m-%Y %H:%M')
```

```python
df_traders['date'] = df_traders['Timestamp IST'].dt.date
```

```python
df_traders['date'] = pd.to_datetime(df_traders['date'])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df_closed = df_traders[df_traders['Closed PnL'] != 0].copy()
```

```python
print(df_closed.shape)
```

```python
df_merged = pd.merge(df_closed, df[['date', 'classification']], on='date', how='left')
```

```python
print(df_merged.shape)
```

```python
df_merged['classification'].value_counts()
```

```python
df_merged['classification'].isnull().sum()
```

```python
df_merged = df_merged.dropna(subset=['classification'])
```

```python
df_merged['sentiment'] = df_merged['classification'].replace({
    'Extreme Fear': 'Fear',
    'Extreme Greed': 'Greed'
})
```

```python
daily = df_merged.groupby(['date', 'Account', 'sentiment']).agg(
    daily_pnl=('Closed PnL', 'sum'),
    trade_count=('Closed PnL', 'count'),
    avg_size_usd=('Size USD', 'mean'),
    win_count=('Closed PnL', lambda x: (x > 0).sum()),
    long_count=('Side', lambda x: (x == 'BUY').sum()),
    short_count=('Side', lambda x: (x == 'SELL').sum())
).reset_index()
```

```python
daily['win_rate'] = daily['win_count'] / daily['trade_count']
```

```python
daily['long_short_ratio'] = daily['long_count'] / (daily['short_count'] + 1)
```

```python
print(daily.shape)
daily.head()
```

## Part A Summary — Data Preparation

- fear_greed_index.csv: 2644 rows, 4 columns, 0 nulls, 0 duplicates
- historical_data.csv: 211,224 rows, 16 columns, 0 nulls, 0 duplicates

- Parsed Timestamp IST, extracted date, merged on date with left join
- 6 unmatched dates dropped post-merge
- Sentiment simplified to 3 buckets: Fear, Greed, Neutral
- Final aggregated daily table: 1,692 rows (one per trader per day)
- Metrics created: daily_pnl, trade_count, avg_size_usd, win_rate, long_short_ratio

## Part B Summary — Analysis Findings

**B1. Performance vs Sentiment**
Greed days show 37% higher median daily PnL than Fear days ($909 vs $663).
However this aggregate result is misleading — when broken down by segment,
High Performers actually earn 67% MORE on Fear days than Greed days.
The overall Greed advantage is driven entirely by Mid Performers.

**B2. Behavior vs Sentiment**
Trade frequency is largely unchanged across sentiment (Fear=20, Neutral=23, Greed=21 trades/day).
Position sizes are slightly larger on Greed days ($1857 vs $1735).
The strongest behavioral signal is long/short ratio — traders are 2x more long-biased
on Greed days (0.27) vs Fear days (0.13), indicating sentiment directly influences directional bias.

**B3. Trader Segmentation**
32 traders segmented into Low/Mid/High Performers by total PnL.
High Performers: median $677K total PnL, 2949 trades, $6920 avg position size.
Mid Performers: median $117K total PnL, 1034 trades, $2429 avg position size.
Low Performers: median $21K total PnL, 1607 trades, $7256 avg position size.
Win rates are nearly identical across segments (83-87%) — volume and sizing drive the gap.

```python
trader_profile = daily.groupby('Account').agg(
    total_pnl=('daily_pnl', 'sum'),
    avg_win_rate=('win_rate', 'mean'),
    total_trades=('trade_count', 'sum'),
    avg_size=('avg_size_usd', 'mean')
).reset_index()

trader_profile['segment'] = pd.qcut(trader_profile['total_pnl'], q=3, labels=['Low Performer', 'Mid Performer', 'High Performer'])

print(trader_profile['segment'].value_counts())
print(trader_profile.groupby('segment', observed=True)[['total_pnl', 'avg_win_rate', 'total_trades', 'avg_size']].median())
```

```python
daily_seg = daily.merge(trader_profile[['Account', 'segment']], on='Account')
seg_sentiment = daily_seg.groupby(['segment', 'sentiment'], observed=True)['daily_pnl'].median().reset_index()
print(seg_sentiment)
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=seg_sentiment, x='segment', y='daily_pnl', hue='sentiment', palette='RdYlGn', ax=ax)
ax.set_title('Median Daily PnL by Trader Segment and Sentiment')
ax.set_xlabel('Trader Segment')
ax.set_ylabel('Median Daily PnL (USD)')
plt.tight_layout()
plt.savefig('segment_sentiment.png', dpi=150)
plt.show()
```

## Part C — Actionable Strategy Recommendations

**Strategy 1: High Performers should lean into Fear days**
High Performer traders generate 67% more daily PnL during Fear vs Greed days ($2168 vs $1294).
During Fear periods, skilled traders should maintain or increase activity rather than pulling back.
Likely explanation: they short panic-driven selloffs or buy undervalued dips while others exit.

**Strategy 2: Mid Performers should wait for Greed confirmation**
Mid Performers earn 6x more on Greed days than Fear days ($592 vs $101).
These traders should reduce position sizes during Fear periods and increase activity
only when sentiment turns Greed — their edge depends on directional momentum, not volatility.

**Strategy 3: Long bias only works in Greed markets**
Long/Short ratio doubles from Fear (0.13) to Greed (0.27) days.
Traders should avoid holding long-biased portfolios during Fear — the data shows
the market rewards short or neutral positioning during those periods.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
features = trader_profile[['avg_win_rate', 'total_trades', 'avg_size', 'total_pnl']].copy()
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
trader_profile['cluster'] = kmeans.fit_predict(features_scaled)
cluster_summary = trader_profile.groupby('cluster')[['avg_win_rate', 'total_trades', 'avg_size', 'total_pnl']].median()
print(cluster_summary)
```

Cluster 1 — Power Traders: highest volume (6279 trades), large positions ($6920), massive PnL ($940K). These are the dominant players.

Cluster 2 — Precision Traders: highest win rate (95.3%), low volume (746 trades), largest positions ($7992), decent PnL ($106K). They trade rarely but very accurately.

Cluster 0 — Active Mid-Tier: middle ground on everything — moderate volume, moderate size, lowest PnL ($88K).

```python
cluster_names = {0: 'Active Mid-Tier', 1: 'Power Traders', 2: 'Precision Traders'}
trader_profile['archetype'] = trader_profile['cluster'].map(cluster_names)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(data=trader_profile, x='total_trades', y='total_pnl', hue='archetype', s=100, ax=axes[0])
axes[0].set_title('Trader Archetypes: Volume vs PnL')
axes[0].set_xlabel('Total Trades')
axes[0].set_ylabel('Total PnL (USD)')
sns.scatterplot(data=trader_profile, x='avg_win_rate', y='total_pnl', hue='archetype', s=100, ax=axes[1])
axes[1].set_title('Trader Archetypes: Win Rate vs PnL')
axes[1].set_xlabel('Average Win Rate')
axes[1].set_ylabel('Total PnL (USD)')
plt.suptitle('Behavioral Archetypes via KMeans Clustering', fontsize=14)
plt.tight_layout()
plt.savefig('clustering.png', dpi=150)
plt.show()
```

## Bonus — Behavioral Archetype Clustering (KMeans)

Three distinct trader archetypes identified via KMeans clustering on win rate, trade volume, position size, and total PnL:

**Power Traders (Cluster 1)** — 11 traders
High volume (6279 median trades), large positions ($6920), dominant PnL ($940K).
These traders win through scale — not necessarily higher accuracy.

**Precision Traders (Cluster 2)** — 11 traders
Highest win rate (95.3%), low volume (746 trades), largest avg position ($7992).
They trade selectively but size up when they do — quality over quantity.

**Active Mid-Tier (Cluster 0)** — 10 traders
Moderate across all metrics. Decent activity but neither the volume of Power Traders
nor the accuracy of Precision Traders. Room to grow in either direction.

Key finding: Win rate alone does not predict PnL. Power Traders have lower win rates
than Precision Traders but 9x higher total PnL — volume and position sizing matter more.
